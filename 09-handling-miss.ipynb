```python
import pandas as pd
```

# missing value handling

```python
d=pd.read_csv('/kaggle/input/sbi-and-sample-data-stock-analysis/SBIN55.csv')
d
```

```python
# removing null value
d.dropna()
```

```python
# removing null value colwise
d.dropna(axis=1)
```

```python
# removing null value rowwise
d.dropna(axis=0)
```

```python
d.dropna(how='any')
# here any is that numll value , like NA , N , NULL , NAN etc
```

```python
d.dropna(subset=['Open'])
```

```python
d.dropna(inplace=True)
d
```

# fillna

```python
d.fillna('python')
```

```python
d.fillna({'Open':'Python'})
```

```python
d.fillna(method='ffill')
```

```python
d.fillna(method='ffill')
```

```python
d.fillna(method='ffill',axis=1) # col
```

```python
d.fillna(method='ffill',axis=0) # row
```

```python
d.fillna(12,inplace=True)
d
```
