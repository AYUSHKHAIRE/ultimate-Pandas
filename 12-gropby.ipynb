```python
import pandas as pd
```

# Groupby

```python
a=pd.DataFrame({"name":["a","b","c","d","a","c","d","b"],"s":[2,3,4,3,54,6,5,6]})
```

```python
a
```

```python
b=a.groupby("name")
b
```

```python
for x,y in b:
    print(x,":",y)
```

```python
b.get_group('a')
```

```python
b.min()
```

```python
b.max()
```

```python
b.mean()
```
