```python
import pandas as pd
```

# handling missing value 

```python
a=pd.read_csv('/kaggle/input/sbi-ana-55/SBIN55.csv')
a
```

```python
a.replace(to_replace=1,value=22)
```

```python
a.replace(to_replace="Open",value="Python")
```

```python
# a.replace(to_replace=[1,100],value=22)
a.replace([1,2,3,4,5,6,7,8,9],22)
```

```python
a.replace("[A-Z,a-z]","python",regex=True)
```

```python
a.replace({"CN":['A-Z']},11,regex=True)
```

```python
a.replace(457.55,method="ffill")
```

```python
a.replace(457.55,method="bfill")
```

```python
a.replace(457,method="ffill",limit=3)
```

```python
a.replace(457,method="ffill",limit=3,inplace=True)
a
```
