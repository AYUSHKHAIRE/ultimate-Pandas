```python
import pandas as pd
```

# Melt()

```python
a = pd.DataFrame({"Days": [1, 2, 3, 4,5,6], "eng": [5, 6, 7, 8,11,23],"Maths":[7,6,7,6,7,6]})
a
```

```python
# reshape
pd.melt(a)
```

```python
pd.melt(a,id_vars=["eng"])
# english works as a number
```

```python
pd.melt(a,id_vars=["Days"],var_name="Python")
```

```python
pd.melt(a,id_vars=["Days"],var_name="Python",value_name="Ayush")
```

# pivot

```python
b = pd.DataFrame({"Days": [1, 2, 3, 4,5,6], "eng": [5, 6, 7, 8,11,23],"Maths":[7,6,7,6,7,6],"Student":['Asd','sdf','dfg','fgh','ghj','hjk']})
b
```

```python
a.pivot(index="Days",columns="Student")
```

```python
a.pivot(index="Days",columns="Student",values="eng")
```

```python
b.pivot_table(index="Student",columns="Days",aggfunc="mean")
```

```python
b.pivot_table(index="Student",columns="Days",aggfunc="sum")
```

```python
b.pivot_table(index="Student",columns="Days",aggfunc="mean",margins=True)
```

```python
b.pivot_table(index="Student",columns="Days",aggfunc="sum",margins=True)
```
