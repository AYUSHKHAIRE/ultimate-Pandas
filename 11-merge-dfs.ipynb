```python
import pandas as pd
```

# merging two dataframe

```python
a=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
a
```

```python
b=pd.DataFrame({"A":[4,3,2,1],"C":[8,7,6,5]})
b
```

```python
pd.merge(a,b,on="A")
```

```python
pd.merge(b,a,on="A")
```

```python
b=pd.DataFrame({"A":[4,3,2,15],"C":[8,7,6,5]})
pd.merge(b,a,on="A")
```

```python
pd.merge(b,a,how="left")
```

```python
pd.merge(b,a,how="right")
```

```python
pd.merge(b,a,how="outer")
```

```python
pd.merge(b,a,how="left",indicator=True)
```

```python
c=pd.DataFrame({"A":[4,3,2,1],"B":[8,7,6,5]})
```

```python
pd.merge(a,c,right_index=True,left_index=True)
```

```python
pd.merge(a,c,right_index=True,left_index=True,suffixes=(" nmae"," id"))
```

# Concating

```python
s1=pd.Series([1,2,3,4])
s2=pd.Series([5,6,7,8])
```

```python
s1
```

```python
s2
```

```python
pd.concat([s1,s2])
```

```python
pd.concat([a,b])
```

```python
pd.concat([a,b],axis=1)
```

```python
pd.concat([a,b],axis=0)
```

```python
pd.concat([a,b],axis=1,join="outer")
```

```python
pd.concat([a,b],axis=1,join="inner")
```

```python
pd.concat([a,b],axis=1,join="inner",keys=["a","b"])
```

```python
pd.concat([a,b],axis=0,join="inner",keys=["a","b"])
```
